```python
import pygame
import random as r
import time
import numpy as np

# вероятность направления прыжка кролика в зависимости от положения
def chance(cord,limit):
    if cord <= limit*0.3:
        choises = 10
    if cord >= limit*0.7:
        choises = -10
    elif limit*0.3 < cord < limit*0.7:
        choises = 0
    return choises

# начальный список координат травы
def cords_grass(limit):
    list_1 = list(np.random.randint(0, limit-size_grass_min, max_grass)/10)
    list_2 = []
    for i in range(max_grass):
        list_2.append(round(list_1[i])*10)
    return list_2

def eat_grass(coord_grass,size_grass, coord_rabbit,size_rabbit):
    list1 = list(np.arange(coord_grass-size_grass,coord_grass+size_grass,0.1)) # списки с учётом размеров, шаг 0.1
    list2 = list(np.arange(coord_rabbit-size_rabbit,coord_rabbit+size_rabbit,0.1))
    for j in range(len(list1)):
        list1[j] = float(f'{list1[j]:.2f}') # округление до 2 знака после запятой
    for j in range(len(list2)):
        list2[j] = float(f'{list2[j]:.2f}')

    for elem_list2 in list2: # есть ли хотя бы кусочек кролика на хотя бы кусочке травы
        if elem_list2 in list1:
            return True # если есть - True
            break
        else:
            return False



pygame.init()

dis_width = 600 # размеры окошка
dis_height  = 600
max_grass = int(dis_width*dis_height*0.0005) # максимальное количество травы
speed_rabbit = 5 # скорость движения кролика

clock = pygame.time.Clock()
dis = pygame.display.set_mode((dis_width,dis_height)) # окошко
pygame.display.set_caption('game?..') # заголовок

# цвета:
sand = (255,235,205)
green = (50,205,50)
black = (0,0,0)
gray = (205,201,201)

# размеры
size_grass_min = 3
size_grass = [size_grass_min]*max_grass # начальный размер всей травы
size_rabbit = 10  # начальный размер кроликов

# координаты
x_grass = cords_grass(dis_width)
y_grass = cords_grass(dis_height)

x_rabbit = 10 # отсчёт с левого верхнего угла
y_rabbit = 10

# помощь в отображении
step_1 = 1
step_2 = 0


game_over = False
while not game_over:

    # позволяет закрыть окно вручную
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            game_over = True
    

    # изменение координат кролика
    x_choises,y_choises = chance(x_rabbit,dis_width),chance(y_rabbit,dis_height)
    x_rabbit_change = r.choice([-10,0,10,x_choises])
    y_rabbit_change = r.choice([-10,0,10,y_choises])
    x_rabbit += x_rabbit_change
    y_rabbit += y_rabbit_change


    # встреча кролика с краем окошка
    while x_rabbit >= dis_width or x_rabbit <= 0 or y_rabbit >= dis_height or y_rabbit <= 0:
        x_rabbit -= x_rabbit_change
        y_rabbit -= y_rabbit_change
        x_choises,y_choises = chance(x_rabbit,dis_width),chance(y_rabbit,dis_height)
        x_rabbit_change = r.choice([-10,0,10,x_choises])
        y_rabbit_change = r.choice([-10,0,10,y_choises])
        x_rabbit += x_rabbit_change
        y_rabbit += y_rabbit_change


    x_now, y_now = x_grass[:step_1], y_grass[:step_1] # координаты отображаемой сейчас травы
    # съедание травы
    for i in range(len(x_now)):
        #if x_rabbit == x_now[i] and y_rabbit == y_now[i]:
        #    size_rabbit += size_grass[i]
        #    x_now[i],y_now[i] = -10,-10
        #    size_grass[i] = 0

        if eat_grass(x_now[i],size_grass[i], x_rabbit,size_rabbit) and eat_grass(y_now[i],size_grass[i], y_rabbit,size_rabbit):
            size_rabbit += size_grass[i]
            x_now[i],y_now[i] = -10,-10
            size_grass[i] = 0



    # фон
    dis.fill(sand) 
    # кролики
    pygame.draw.rect(dis,gray,[x_rabbit,y_rabbit,size_rabbit,size_rabbit])

    # трава
    for i in range(len(x_now)):
        # изменение размера травы
        if size_grass[i-1] <= 10:
            size_grass[i-1] += 0.1
        # умирание травы
        if step_1 >= max_grass:
            size_grass[step_2] = size_grass_min
            x_grass[step_2] = round(r.randint(0, dis_width-size_grass_min)/10)*10
            y_grass[step_2] = round(r.randint(0, dis_width-size_grass_min)/10)*10
        
        pygame.draw.rect(dis,green,[x_now[i],y_now[i],size_grass[i],size_grass[i]])
    
    
    step_1 += 1
    step_2 += 1
    if step_2 >= len(x_grass):
        step_2 = 0

    pygame.display.update()
    clock.tick(speed_rabbit)

    
pygame.display.update()
time.sleep(1)
pygame.quit()
quit()
```
